# Fault injection data from Tesla K20

On Kepler, it is possible to inject fault on SASSIFI and NVBITFI, and both results are available on data/ directory.

```python
import pandas as pd

#load sassifi
sassifi_df = pd.read_csv("data/avf_kepler_sassifi.csv", index_col=[0,1])

#load nvbitfi
nvbitfi_df = pd.read_csv("data/avf_kepler_nvbitfi.csv", index_col=[0,1])
```

```python
# Define a function to plot both graphs
import matplotlib.pyplot as plt
%matplotlib inline

def plot(df):
    fig = plt.figure()
    ax = fig.add_subplot(111)

    df.plot.bar(edgecolor="black", colormap="RdYlBu", ax=fig.gca(), stacked=True)
    plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
```

```python
# Group to plot as in the paper
sassifi_df.index.names = ["outcome", "igid"]
sassifi_df = sassifi_df.groupby(["outcome"]).sum()

# add masked
sassifi_df.loc["Masked"] = 1 - sassifi_df.sum(axis=0)

#Plot
plot(sassifi_df.transpose())
```

```python
# Group to plot as in the paper for NVBITFI
nvbitfi_df.index.names = ["outcome", "igid"]
nvbitfi_df = nvbitfi_df.groupby(["outcome"]).sum()

# add masked
nvbitfi_df.loc["Masked"] = 1 - nvbitfi_df.sum(axis=0)

#Plot
plot(nvbitfi_df.transpose())
```
